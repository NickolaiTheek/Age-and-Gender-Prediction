# Part B – Compare and evaluate AI techniques.

 - Object recognition using Decision Trees, Random Forest and CNN with Regression .

 #### 1. Decision trees

  Among the most powerful ensemble learning methods, Random Forests are ensembles of several decision trees that boost the prediction accuracy and reduce overfitting. They are very robust and versatile since they do not assume any specific distribution of data, hence turning them effective in both classification and regression tasks. Some of the salient features of Random Forests are the ability to give insight into feature importance, thus showing which attributes contribute most to predictions. On the other hand, Random Forests can be used to predict age and gender by handling imbalanced data through balanced sampling during training. However, it requires pre-processed numerical features since it cannot process raw images. For example, it would be necessary to extract in advance the features representing the facial characteristics, such as distances between landmarks or averages of pixel intensities.

  While Random Forests are excellent in the case of structured, tabular data, relying on feature extraction does not make them ideal for raw image-based tasks. While this may be so, training a large number of trees is computationally expensive, and by nature, an ensemble of trees is less interpretable than a single decision tree. Despite these limitations and challenges, Random Forests can give sturdy predictions of age as a continuous variable and gender as a classification task, given meaningful features extracted from the images. It is thus suited to scenarios where computational efficiency and flexibility are in view, provided sufficient feature engineering is done to prepare the data for this.

 #### 2. Random Forest

  Decision Trees represent relatively simple, transparent machine learning that works rather well when the data are structured. Simplicity is enclosed in a tree-like structure; hence so simple to visualize and understand. Considering an age and gender prediction problem, Decision Trees will work if some pre-engineered numerical features, such as facial metrics or intensity averages, are used. They can support nonlinear relationships in the data, making them quite versatile for a wide variety of structured datasets. They are also rather inexpensive in computational cost, which makes them attractive in cases of smaller datasets or limited computational resources.

  However, Decision Trees have a number of remarkable weaknesses: they might suffer from overfitting, especially if deep, generally performing worse in comparison with ensemble models like Random Forests or deep learning approaches such as CNNs when it comes to complex tasks or big datasets. They are not fit for raw image data processing either, since feature extraction is a must in that case. Besides, decision trees are sensitive to noisy data that can give unreliable predictions. Despite these drawbacks, decision trees can be viable and light solutions for age and gender predictors if meaningful features are extracted and the problem of imbalanced datasets is treated properly.

  #### 3. CNNs and regression

  The CNN is great, therefore, for handling image-based applications and serves as the best avenue for these age and gender predictions. Since CNNs, on their own, extract features from raw images due to patterns and the structure of faces, they will process complex information in a raw form accurately only in large-scale datasets with explicit feature engineering. This flexibility in their end-to-end learning framework lets them be very useful for both classification tasks-say, like gender prediction-and regression tasks like the age estimation of a face. Still, CNNs are computationally intensive and require enormous hardware resources in the form of GPU resources for training and inference. They also show tendencies to overfit to small datasets, which regularization and proper augmentation should improve upon for generalization. On the other hand, regression is lightweight, apt, and utilized in the case of predicting continuous variables such as age.
  
  Regression provides computational efficiency and interpretability where clear relations between the input features and the output predictions can be clearly provided. However, no regression model can operate directly on the raw images and requires pre-engineered features, such as distances between facial landmarks or averages of pixel intensities. While regression does a fairly good job on structured datasets, due to their very nature, assumptions regarding linearity or pre-defined relationships make it difficult for them to capture the complexity in image-based tasks. While your model can directly carry out age and gender predictions in raw face images using CNNs, regarding regression, the features first need to be extracted from such images in numeric formats. While these CNNs promise more flexibility, hence higher accuracy, the regression models favor higher computational resources or small dataset sizes.

- Processes the dataset of facial images, extracting valid file paths and labels for age, gender, and ethnicity while filtering out invalid or irrelevant files.

```python
import os
from tqdm import tqdm  # The 'tqdm' library is used to show a progress bar for the loop.

# Initializing empty lists to store image paths and their corresponding labels for age and gender.
image_paths = []
age_labels = []
gender_labels = []


for filename in tqdm(os.listdir(BASE_DIR)):
    # Skipping hidden files
    if filename.startswith('.') or not filename.lower().endswith('.jpg'):
        continue

    # Splitting the filename into its name and extension parts
    name, ext = os.path.splitext(filename)
    parts = name.split('_')

    if len(parts) < 4:
        continue

    try:
        # Extracting the age, gender, and ethnicity labels from the filename.
        age = int(parts[0])
        gender = int(parts[1])
        ethnicity = int(parts[2])
    except ValueError:

        continue

    # Adding the valid image file path and its corresponding labels to the respective lists.
    image_paths.append(os.path.join(BASE_DIR, filename))
    age_labels.append(age)
    gender_labels.append(gender)
```

```python
# convert to dataframe
df = pd.DataFrame()
df['image'], df['age'], df['gender'] = image_paths, age_labels, gender_labels
df.head()
```

```python
# map labels for gender
gender_dict = {0:'Male', 1:'Female'}
```

- Ensures that all images are uniformly preprocessed for consistent model input.

```python
def extract_features(images):
    features = []
    for image in tqdm(images):
        img = load_img(image, color_mode='grayscale')
        img = img.resize((128, 128), Image.Resampling.LANCZOS)
        img = np.array(img)
        features.append(img)

    features = np.array(features)

    features = features.reshape(len(features), 128, 128, 1)
    return features
```

```python
# Prepare the feature array (X) by processing images
X = extract_features(df['image'])
```

```python
# normalize the images
X = X/255.0
```

- Split into Training and Testing Data

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Convert labels to NumPy arrays
y_gender = np.array(gender_labels)
y_age = np.array(age_labels)

# Split the data into training and testing sets
X_train, X_test, y_gender_train, y_gender_test, y_age_train, y_age_test = train_test_split(
    X, y_gender, y_age, test_size=0.2, random_state=42
)

# Check the shapes of the resulting datasets
print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
print("Training labels shape (Gender, Age):", y_gender_train.shape, y_age_train.shape)
print("Testing labels shape (Gender, Age):", y_gender_test.shape, y_age_test.shape)
```

- Train the Model

```python
from keras.models import Model
from keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D, Input, BatchNormalization
from keras.optimizers import Adam
from keras.regularizers import l2
from keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Define the input shape
input_shape = (128, 128, 1)  # Adjust if using grayscale (1 channel) or RGB (3 channels)

# Define the model
inputs = Input((input_shape))

# Convolutional layers with Batch Normalization and L2 Regularization
conv_1 = Conv2D(32, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.001))(inputs)
bn_1 = BatchNormalization()(conv_1)
maxp_1 = MaxPooling2D(pool_size=(2, 2))(bn_1)

conv_2 = Conv2D(64, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.001))(maxp_1)
bn_2 = BatchNormalization()(conv_2)
maxp_2 = MaxPooling2D(pool_size=(2, 2))(bn_2)

conv_3 = Conv2D(128, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.001))(maxp_2)
bn_3 = BatchNormalization()(conv_3)
maxp_3 = MaxPooling2D(pool_size=(2, 2))(bn_3)

conv_4 = Conv2D(256, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.001))(maxp_3)
bn_4 = BatchNormalization()(conv_4)
maxp_4 = MaxPooling2D(pool_size=(2, 2))(bn_4)

# Flatten
flatten = Flatten()(maxp_4)

# Fully connected layers with Dropout and L2 Regularization
dense_1 = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(flatten)
dropout_1 = Dropout(0.3)(dense_1)

dense_2 = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(dropout_1)
dropout_2 = Dropout(0.3)(dense_2)

# Output layers
output_1 = Dense(1, activation='sigmoid', name='gender_out')(dropout_2)  # Gender classification
output_2 = Dense(1, activation='relu', name='age_out')(dropout_2)        # Age regression

# Compile the model with a reduced learning rate
model = Model(inputs=[inputs], outputs=[output_1, output_2])
model.compile(
    loss=['binary_crossentropy', 'mae'],
    optimizer=Adam(learning_rate=0.0001),
    metrics={'gender_out': 'accuracy', 'age_out': 'mae'}
)

# Add callbacks for training
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

#reduces the learning rate of the optimizer when the validation loss stops improving.
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6,
    verbose=1
)

# Train the model
history = model.fit(
    x=X_train,
    y=[y_gender_train, y_age_train],
    batch_size=32,
    epochs=50,
    validation_split=0.2,
    callbacks=[early_stopping, reduce_lr]
)

```

- Visualization of Training and Validation Metrics

```python
import matplotlib.pyplot as plt

# Plot Training and Validation Loss
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss vs. Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot Training and Validation Accuracy for Gender Prediction
plt.figure(figsize=(12, 6))
plt.plot(history.history['gender_out_accuracy'], label='Training Accuracy (Gender)')
plt.plot(history.history['val_gender_out_accuracy'], label='Validation Accuracy (Gender)')
plt.title('Accuracy vs. Epochs (Gender)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot Training and Validation MAE for Age Prediction
plt.figure(figsize=(12, 6))
plt.plot(history.history['age_out_mae'], label='Training MAE (Age)')
plt.plot(history.history['val_age_out_mae'], label='Validation MAE (Age)')
plt.title('MAE vs. Epochs (Age)')
plt.xlabel('Epochs')
plt.ylabel('MAE')
plt.legend()
plt.show()
```

```python
# Evaluate the model on the test set
test_loss, gender_loss, age_loss, gender_accuracy, age_mae = model.evaluate(
    X_test, [y_gender_test, y_age_test], verbose=1
)

# Print the results
print("Test Gender Loss:", gender_loss)
print("Test Age Loss:", age_loss)
print("Test Gender Accuracy:", gender_accuracy)
print("Test Age MAE:", age_mae)
```

# Part D – Testing.
The chosen testing methods for the implementation are listed below
 - Metrics Evaluation
    - Accuracy
    - Precision
    - Recall
    - F1-score
    - Receiver Operating Characteristic (ROC) curve
    - Confusion Matrix
    

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Predictions for gender classification
y_gender_proba = model.predict(X_test)[0].flatten()  # Probabilities
y_gender_pred = (y_gender_proba > 0.5).astype(int)   # Convert probabilities to binary predictions

# Classification Report
print("Classification Report for Gender Prediction (Test Set):")
print(classification_report(y_gender_test, y_gender_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_gender_test, y_gender_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Female', 'Male'], yticklabels=['Female', 'Male'])
plt.title("Confusion Matrix for Gender Prediction")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_gender_test, y_gender_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Diagonal line
plt.title("ROC Curve for Gender Prediction")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.show()

# Summary Table (Accuracy, Precision, Recall, F1, AUC)
accuracy = np.mean(y_gender_test == y_gender_pred)
precision = np.sum((y_gender_pred == 1) & (y_gender_test == 1)) / np.sum(y_gender_pred == 1)
recall = np.sum((y_gender_pred == 1) & (y_gender_test == 1)) / np.sum(y_gender_test == 1)
f1 = 2 * (precision * recall) / (precision + recall)

print("\nSummary Table:")
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Precision: {precision * 100:.2f}%")
print(f"Recall: {recall * 100:.2f}%")
print(f"F1-Score: {f1 * 100:.2f}%")
print(f"AUC: {roc_auc * 100:.2f}%")
```
